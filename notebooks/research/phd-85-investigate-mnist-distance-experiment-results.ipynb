### 2.3. Latent space distance metrics

```python
latent_distance_metrics = [
    'h_norm_ratio',
    'h_true_norm',
    'h_approx_norm',
    'h_norm_classwise',
    # 'h_norm_zeros_ratio',
    # 'h_norm_ones_ratio',
    'h_norm_directionwise',
    # 'h_norm_direction_zeros_ratio',
    # 'h_norm_direction_ones_ratio'
]
```

```python
df[latent_distance_metrics]
```

```python
df[['h_norm_ratio', 'h_norm_classwise', 'h_norm_directionwise']].plot()
```

```python
df[['h_true_norm', 'h_approx_norm']].plot()
```

### Key takeaways. 
The norm ratios are flat around 1. Why?
- True and approximate h-spaces are roughly equal.
- The approximation is good, because it is refit at every iteration.

**The absolute values decrease with OOD.** Why?
- The OOD points are in the middle of the space whereas in-distribution are pushed out to the extremes. So as more are added the latent space shrinks.


### 2.4. Relative latent distance metrics
Relative to validation set

```python
validation_distance_metrics = [
    'validation_h_norm_ratio',
#     'validation_h_true_norm',
#     'validation_h_approx_norm', 
    'validation_h_norm_classwise',
#     'validation_h_norm_zeros_ratio', 
#     'validation_h_norm_ones_ratio',
    'validation_h_norm_directionwise',
#     'validation_h_norm_direction_zeros_ratio',
#     'validation_h_norm_direction_ones_ratio'
]
```

```python
df[validation_distance_metrics]
```

```python
df[['validation_h_norm_ratio', 'validation_h_norm_classwise']].plot()
```

### Key takeaway.
Norms relative to validation set decrease with OOD. Why?
- Same as above, the more OOD points, the more points in the middle of the latent space. So the hypervolume is less than when points are in-distribution and pushed out ot the extremes.
- **This norm is nicely scaled** - capped around 1 for full in-distribution.
- Using a validation set as a reference point resembles the calibration set used by Papernot and Mcdaniel.

```python
df[['validation_h_norm_directionwise']].plot()
```

### Key takeaway. 
When we isolate the noise axis, the norm increases with OOD. Why?
- The in-distribution points all lie in-plane / on the axis, so have small value in the noise axis.
- The OOD points are scattered in the noise axis, so their norm is greater.
